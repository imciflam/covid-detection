# Loss learning curve (bar chart)

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7']
#models = ['LeNet original', 'LeNet modified', 'Model I', 'Model II', 'VGG19', 'Inception v3', 'Xception']

#           LeNet Ori.,     LeNet mod.,   Model I,   Model II,   VGG19,    Inception v3,  Xception
loss_200 = [6.8e-1,         1.75e-7,      7.05e-1,   3.85e-6,    5.65e-7,  1.56e-3,       1.87e-1]
loss_256 = [6.73e-1,        4.47e-9,      6.61e-1,   1.08e-5,    3.04e-3,  4.18e-3,       5.42e-2]
loss_512 = [6.72e-1,        4.99e-8,      6.71e-1,   1.24e-5,    2.12e-2,  2.66e-2,       7.10e-3]

labels = ['200x200', '256x256', '512x512']

x_axis = np.arange(len(models)) * 4

plt.figure(figsize=(9, 4))
plt.rcParams['font.family'] = "Liberation Serif"
plt.rc('font', size=20, weight='bold') #controls default text size
#plt.rc('axes', titlesize=20) #fontsize of the title
plt.rc('axes', labelsize=24) #fontsize of the x and y labels
plt.rc('xtick', labelsize=20) #fontsize of the x tick labels
plt.rc('ytick', labelsize=20) #fontsize of the y tick labels
plt.rc('legend', fontsize=20) #fontsize of the legend
logEnable=True

# Multi bar char
width=1
plt.bar(x_axis - (width) , loss_200, width, label = labels[0], log=logEnable, color='r')
plt.bar(x_axis , loss_256, width, label = labels[1], log=logEnable, color='b')
plt.bar(x_axis + (width), loss_512, width, label = labels[2], log=logEnable, color='g')


# Xticks
plt.xticks(x_axis, models, rotation=0)

# Add legend
plt.legend(['200x200', '256x256', '512x512'], loc='best')

plt.ylabel('Loss', weight='bold')
plt.xlabel('Architecture', weight='bold')

plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('barchart_loss.svg', dpi=150)

# Display
plt.show()
```

# Accuracy learning curve (bar chart)

```python
models = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7']
#models = ['LeNet original', 'LeNet modified', 'Model I', 'Model II', 'VGG19', 'Inception v3', 'Xception']


#          LeNet Ori.,     LeNet mod.,   Model I,   Model II,   VGG19,    Inception v3,  Xception
acc_200 = [5.71e-1,         1,           5.71e-1,   1,          1,        1,             9.36e-1]
acc_256 = [6.08e-1,         1,           6.08e-1,   1,          9.97e-1,  1,             9.83e-1]
acc_512 = [6.08e-1,         1,           8.50e-1,   1,          9.98e-1,  1,             1]

labels = ['200x200', '256x256', '512x512']

x_axis = np.arange(len(models)) * 4

plt.figure(figsize=(9, 4))
plt.rcParams['font.family'] = "Liberation Serif"
plt.rc('font', size=20, weight='bold') #controls default text size
#plt.rc('axes', titlesize=20) #fontsize of the title
plt.rc('axes', labelsize=24) #fontsize of the x and y labels
plt.rc('xtick', labelsize=20) #fontsize of the x tick labels
plt.rc('ytick', labelsize=20) #fontsize of the y tick labels
plt.rc('legend', fontsize=20) #fontsize of the legend
logEnable=False

# Multi bar char
width=1
plt.bar(x_axis - (width) , acc_200, width, label = labels[0], log=logEnable, color='r')
plt.bar(x_axis , acc_256, width, label = labels[1], log=logEnable, color='b')
plt.bar(x_axis + (width), acc_512, width, label = labels[2], log=logEnable, color='g')


# Xticks
plt.xticks(x_axis, models, rotation=0)

# Add legend
plt.legend(['200x200', '256x256', '512x512'], loc='best')

plt.ylabel('Accuracy', weight='bold')
plt.xlabel('Architecture', weight='bold')

plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('barchart_accuracy.svg', dpi=150)

# Display
plt.show()
```
